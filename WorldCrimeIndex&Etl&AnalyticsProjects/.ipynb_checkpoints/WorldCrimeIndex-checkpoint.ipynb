```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import sqlite3
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
""" That's gonna create the logging function."""
log_file = "log_file.txt" 
def log_progress(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S'
    now = datetime.now()
    timestamp = now.strftime(timestamp_format)
    with open(log_file, "a") as f:
        f.write(timestamp + ',' + message + '\n')
```

```python
# EXTRACT-1
"""This block, extract the table-data from wikipedia."""
log_progress("ETL Job Started") 
log_progress("EXTRACT PHASE STARTED")
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_intentional_homicide_rate'

conn = sqlite3.connect('STAFF.db')
table_name = 'WorldCrimeIndex'
attribute_list = ["LOCATION","RATE","COUNT","YEAR","REGION","SUBREGION"]
csv_path = 'CrimeIndex.csv'
df = pd.DataFrame(columns=["Location","Rate","Count","Year","Region","Subregion"])

html_page = requests.get(url).text
data = BeautifulSoup(html_page, 'html.parser')

tables = data.find_all('tbody')
rows = tables[1].find_all('tr')
```

```python
# EXTRACT-2
"""This block, create a dataframe that includes extracted table-data from wikipedia."""
count=0
for row in rows:
    if count<len(rows):
        col = row.find_all('td')
        if len(col)!=0:
            data_dict = {"Location": col[0].get_text(strip=True),
                         "Rate": float(col[1].get_text(strip=True).replace(',','')),
                         "Count": int(col[2].get_text(strip=True).replace(',','')),
                         "Year": int(col[3].get_text(strip=True)),
                         "Region": col[4].get_text(strip=True),
                         "Subregion": col[5].get_text(strip=True)
                         }
            df1 = pd.DataFrame(data_dict, index=[0])
            df = pd.concat([df,df1], ignore_index=True)
            count+=1
    else:
        break
log_progress("EXTRACT PHASE ENDED")
```

```python
df
```

```python
#TRANSFORM
"""Some countries have '*' character at the end of the string. I removed this '*'."""
log_progress("TRANSFORM PHASE STARTED")
df['Location'] = df['Location'].str.replace('*', '')
log_progress("TRANSFORM PHASE ENDED")
```

```python
df
```

```python
### LOAD
"""Finally, We can load our data to the database and the csv."""
log_progress("LOAD PHASE STARTED")
df.to_csv(csv_path)
df.to_sql(table_name, conn, if_exists = 'replace', index=False)
log_progress("LOAD PHASE ENDED")
```

```python
# LOGGING ENDED
"""ETL JOB ENDED"""
log_progress("ETL Job Ended") 
```

```python
"""We wrote an SQL query that displays the total count and average rate of all murders by Subregions."""
query_statement = f"SELECT SubRegion,sum(Count) as Count, avg(Rate) as Rate from {table_name} Group By Subregion"
query_output = pd.read_sql(query_statement, conn)
print(query_statement)
df_World_Subregion = pd.DataFrame(query_output)
df_World_Subregion
```

```python
"""We wrote an other SQL query that displays the total count and average rate of all murders by Regions"""
query_statement2 = f"SELECT Region,sum(Count) as Count,avg(Rate) as Rate from {table_name} Group By Region"
query_output2 = pd.read_sql(query_statement2, conn)
print(query_statement2)
df_World_Region = pd.DataFrame(query_output2)
df_World_Region
```

```python
"""We wrote an other SQL query that displays the total count and average rate of all murders by Countries."""
query_statement3 = f"SELECT Location,max(Count) as Count, Rate from {table_name} where count>1000 group by Location order by Count desc"
query_output3 = pd.read_sql(query_statement3, conn)
print(query_statement3)
df_WorldCount = pd.DataFrame(query_output3)
df_WorldCount.head()
```

```python
"""We wrote an other SQL query that displays the average rate of all murders by Countries."""
query_statement4 = f"SELECT Location, Count,Rate from {table_name} where count>1000 group by Location order by Rate desc"
query_output4 = pd.read_sql(query_statement4, conn)
print(query_statement4)
df_WorldRate = pd.DataFrame(query_output4)
df_WorldRate.head()
```
